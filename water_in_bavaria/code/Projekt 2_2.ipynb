```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Wasserhärte_Bayern')
del df['Unnamed: 0']
del df['water_hardness_mmol_l']
del df['water_hardness_fH']
df
```

```python
df['water_hardness_dH'] = pd.to_numeric(df['water_hardness_dH'], errors='coerce')
```

```python
df['water_hardness_dH'] = df.groupby('zip_code')['water_hardness_dH'].transform('mean')
```

```python
df
```

```python
conditions = [
    df['water_hardness_dH'] <= 8.4,
    (df['water_hardness_dH'] > 8.4) & (df['water_hardness_dH'] <= 14),
    (df['water_hardness_dH'] > 14) & (df['water_hardness_dH'] <= 21)
]
values = ['weich', 'mittel', 'hart']

df['hardness_range'] = np.select(conditions, values, default=None)

df
```

```python
df1 = df.drop_duplicates(subset='zip_code').reset_index(drop=True)
df2 = df1.dropna()
final_df = df2.sort_values('zip_code').reset_index(drop=True)
final_df
```

```python
final_df[['latitude', 'longitude']] = final_df['coordinates'].str.split(' / ', expand=True)

final_df
```

```python
final_df.to_csv('Wasserdaten_Bayern.csv')
```
